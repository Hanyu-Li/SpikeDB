```python
def downsample(c):
    # final spike removal
    
    plot(c)
    show()
    end=c.size-1
    #find last spike
    window = 10000
    threshold = 0.15
    last_max = c[end]
    while end > window:
        max_v=c[end-window:end].max()
        print max_v
        if max_v - last_max>threshold:
            print 'spike'
            break;
        else:        
            last_max = max_v;
            
        end -= window
    
    print end   
    try:    
        c[end-window:end+window] = c[end+window:end+3*window]
    except:
        print 'final spike not removed'
     
    plot(c)
    show()
    c = c/c.max()
    d = c[0:c.size:1000]
    print c
    plot(d)
    show()
    return d
```

```python
r = collection.find_one({'fnhead' :'2011_03_09_', 'fnum':95})
#r = collection.find_one()
c = SpikeDB.read_data(r, 'c')
d = downsample(c)
plot(d)
show()
```
